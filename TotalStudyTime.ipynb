```python
import pandas as pd
import numpy as np
```

```python
FILE_NAME = "data/carga_horaria.csv"
data = pd.read_csv(FILE_NAME, sep=';')
```

```python
data
```

```python
data = data.drop(10)
```

```python
data
```

```python
data_total_horas_de_estudo = data["Total Horas de Estudo"]
data.loc[data_total_horas_de_estudo > 0, ["Total Horas de Estudo"] ] = 0
data
```

```python
import math
HORAS_POR_DIA_ESTUDO = 6
DIAS_DA_SEMANA_ESTUDO = 6

total_dificuldade = sum(data["dificuldade"])
for i in range(len(data["dificuldade"])):
	total_horas_de_estudo = math.ceil(((HORAS_POR_DIA_ESTUDO * DIAS_DA_SEMANA_ESTUDO)/ total_dificuldade) * data["dificuldade"][i])
	data_total_horas_de_estudo[i] = format(total_horas_de_estudo, '.0f')
data
```

```python
new_data = data.copy()
new_data = new_data.drop(columns=["dificuldade", "Quant. De Questões"])

new_data.to_csv('data/total_study_time.csv', index=False)
```
